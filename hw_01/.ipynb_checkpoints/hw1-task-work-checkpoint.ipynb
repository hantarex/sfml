```python
import pandas as pd
%matplotlib inline
```

```python
# (1) Используя параметры pandas прочитать красиво пандас 
df = pd.read_csv("UCI_Credit_Card.csv")
```

```python
df.head()
```

```python
grap = df.pivot_table(index=['EDUCATION','SEX'], values='LIMIT_BAL')
```

```python
grap.unstack("SEX")
```

```python
import matplotlib.pyplot as plt
_, ax = plt.subplots(figsize=(20,10))
grap.unstack("SEX").plot(kind="bar", title="Зависимость кредитного лимита от образования для каждого пола",ax=ax)
ax.legend(("MAN", "WOMAN"), title="SEX", loc=1);
plt.show()
```

```python
import seaborn as sns
man = df[df['SEX'] == 1]
ax = sns.factorplot('EDUCATION','LIMIT_BAL',data=man, size=10)
ax.fig.suptitle("Зависимость кредитного лимита от образования только для мужчин")
```
